## 3.4. Focus on the problem (good employee left prematurely)

Above correlation is based on the whole employees. As the question is actually asking why the best employees are leaving, we should classify the data based on performance. 

So how do we define a employee as good? Simply by the evaluation?

First let's classified the evaluation into 10 groups.

```python
#add mapping for left and stay
left_dic = {1:'left', 0:'stay'}
data['left_str'] = data['left'].map(left_dic)

#define function for evaluation classification
def eva_group(x):
    for i in np.arange(1,11):
        if x > (i-1)/10 and x <= i/10:
            return i

data['eva_group_int'] = data['last_evaluation'].apply(lambda x: eva_group(x))
eva_data = data[['eva_group_int','left_str','left']].groupby(['eva_group_int','left_str']).count()
eva_data = eva_data.reset_index()
eva_data
```

```python
#as no employee left with evaluation 4, we fill it manually
eva_data.loc[13] = [4,'left',0]
eva_data = eva_data.sort_values(by='eva_group_int')
eva_data
```

```python
eva_data_left = eva_data[eva_data['left_str'] == 'left']
eva_data_stay = eva_data[eva_data['left_str'] == 'stay']
#count of stay, for horizontal and vertical
eva_stay_h = eva_data_stay['eva_group_int'].tolist()
eva_stay_v = eva_data_stay['left'].tolist()
#count of left, for horizontal and vertical
eva_left_h = eva_data_left['eva_group_int'].tolist()
eva_left_v = eva_data_left['left'].tolist()

#add them to the same bar, to make it easy to see how may employees left in each evaluation group
f, ax = plt.subplots(figsize=(10,5))
ax.bar(eva_stay_h, eva_stay_v, label='stay', bottom=eva_left_v)
ax.bar(eva_left_h, eva_left_v, label='left')
plt.legend()
plt.show()
```

The bar plot shows employees that fall into evaluation group 6 to 7 and 9 to 10 have a high turnover rate. But should we take only group 10 as good or both 9 and 10?

Ronald Troncoso defined a good employee as:
***"anyone who's performance evaluation was at least 1 standard deviation above the mean(Top 84th percentile)"*
**
in his [post][1].
This is actually the usage of [z-score][2](Standard Score). He define good employee as z-score >= 1. 


  [1]: https://www.kaggle.com/ronaldtroncoso20/d/ludobenistant/hr-analytics/hr-analytics-why-are-so-many-employees-leaving
  [2]: https://en.wikipedia.org/wiki/Standard_score

```python
#calculate according to z-score
eva_mean = data['last_evaluation'].mean()
eva_std = data['last_evaluation'].std()
data['eva_label'] = (data['last_evaluation'] - eva_mean)/eva_std

def eva_label(x):
    if x>1:
        return 'good'
    else:
        return 'normal'
    
data['eva_label'] = data['eva_label'].apply(lambda x: eva_label(x))
eva_data_zscore = data[['eva_label','left_str','left']].groupby(['eva_label','left_str']).count().reset_index()
eva_data_zscore
```

```python
#compare of two definitions
print(eva_data_zscore)
print("=======================================")
print(eva_data)
```

So if we simply define evaluation group 10 as good employees, we have a small gap with Ronald Troncoso's definition. 

If we take both group 9 and 10 into consideration, the gap will be large. The definition of good employee will affect our analysis result greatly.

As Ronald Troncoso's definition is much more scientific, I will take his definition here.

To make our analysis more directed, we need to identify the employees who left prematurely. This will help us eliminate the influence caused by employee who left after a long service in the company. I will use "time_spend_company" to judge it.  

*Actually, if the company has a history of about 10 years and they define leave prematurely as leave after working for like 5 years or less, then it's meaningful to define the limit of "prematurely". However, if the company has a history of 20 or more years and they take all staffs that leave within 10 years as "prematurely", then there's no need to define the limit here. As we only got 10 years data. This is ambiguous here. Let's assume that the company has a short history.*

```python
#left & stay over the time spend in company
f, ax = plt.subplots(figsize=(12,7))
time_data = data[['time_spend_company','left_str','left']].groupby(['time_spend_company','left_str']).count().reset_index()
sns.barplot(x='time_spend_company', y='left', hue='left_str', data=time_data)
plt.legend(loc='upper right')
```

For employees who stay with company for 5 years, more than 50% of them choose to leave. 5-years is a peak. So let's try to define 5 as the limit of "prematurely". That means, we labeled employee who left the company before the 5th year as "leave prematurely".

```python
def pre_group(x):
    if x<5:
        return 'prema'

#we only label those good employees who left
data['pre_label'] = data[(data['left_str']=='left') & (data['eva_label']=='good')]['time_spend_company'].apply(lambda x: pre_group(x))
data['pre_label'] = data['pre_label'].fillna('z')
data['pre_label'].value_counts()
```

```python
#num_cols = data._get_numeric_data().columns
data['total_label'] = data['left_str']+":"+data['eva_label']+","+data['pre_label']
data.groupby(['total_label']).mean().transpose().drop(['left','eva_group_int','dept_id'])
```

Let's use group 1,2,3,4 and 5 represent the "left-good-prematurely", "left-good", "left-normal", "stay-good" and "stay-normal" groups. And we mainly focus on the difference between group 1 and the others.

 **- The satisfaction level of group 1 is much lower than the other groups. Satisfaction matters.**

 **- Group 1 have high average monthly hours and completed more projects than others. Monthly hours and projects matters.**

 **- Group 1 got no promotion.**

Beside above observation, we also know that the average salary of left group is lower than stay group. Also they have little work accidents compare with stay group.

```python
#good employee
group1_data = data[data['eva_label'] == 'good']
group1_corr = group1_data.corr()
group1 = pd.DataFrame(group1_corr['left'].drop(['left','dept_id']).sort_values())
group1
```
